# Normal Equation Approach

```python
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100,1)
```

```python
plt.scatter(X,y, c = 'blue', marker = '.')
plt.xlabel('$x_1$')
plt.ylabel('$y$')
plt.axis([0, 2, 0, 15])
plt.show()
```

```python
X_b = np.c_[np.ones((100,1)), X]
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python
theta_best
```

```python
# The function that we used to generate the data is, y = 4 + 3x + Gaussian noise
# Close enough, but the noise made it impossible to recover the exact parameters of the original function.
```

```python
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2,1)), X_new]
y_predict = X_new_b.dot(theta_best)
y_predict
```

```python
plt.plot(X_new, y_predict, linestyle = '--', color = 'red', label = 'predictions')
plt.scatter(X,y, c = 'blue', marker = '.')
plt.xlabel('$x_1$')
plt.ylabel('$y$')
plt.axis([0, 2, 0, 15])
plt.legend()
plt.show()
```

```python
# Performing linear regression using scikit - learn is simple.

from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
lin_reg.intercept_, lin_reg.coef_
```

# Batch Gradient Descent

```python
eta = 0.1 # learning rate
n_iterations = 1000
m = 100

theta = np.random.randn(2, 1)

for iteration in range(n_iterations):
    gradients = (2/m) * X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - eta * gradients
```

```python
theta
```

```python
# This is exactly what the normal equation found.
```

# Stochastic Gradient Descent

```python
from sklearn. linear_model import SGDRegressor

sgd_reg = SGDRegressor(max_iter = 1000, tol = 0.001, eta0 = 0.1, penalty = None)
sgd_reg.fit(X, y.ravel())    # y.ravel() converts the 2-D array into a 1-D array
```

```python
sgd_reg.intercept_, sgd_reg.coef_  # The paramters are close to that obtained from the Normal Eqn. / Batch GD method.
```

# Polynomial Regression

```python
m = 100
X = 6 * np.random.rand(m,1) - 3
y = 2 + X + 0.5 * (X**2) + np.random.randn(m,1)
```

```python
plt.scatter(X, y, marker = '.', color = 'blue')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree = 2, include_bias = False)
X_poly = poly_features.fit_transform(X)
X[0]
```

```python
X_poly[0]
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)
lin_reg.intercept_, lin_reg.coef_
```

```python
plt.scatter(X, y, marker = '.', color = 'blue')
plt.xlabel('$x$')
plt.ylabel('$y$')

X_new = np.linspace(-3, 3, 1000).reshape(1000, 1)
X_new_poly = poly_features.transform(X_new)
y_new = lin_reg.predict(X_new_poly)

plt.plot(X_new, y_new, linestyle = '-', linewidth = 2, color = 'red', label = 'predictions')
plt.legend()
plt.show()
```

```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
X = iris['data'][:, 3:] # petal width
X.ndim
```

```python
print(iris['target'])
print(type(iris['target'][0]))
```

```python
y = (iris['target'] == 2).astype(np.int) # 1 if iris virginica, else 0
```

Note: To be future-proof we set solver="lbfgs" since this will be the default value in Scikit-Learn 0.22.

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(solver = 'lbfgs', random_state = 42)
log_reg.fit(X, y)
```

```python
# Let's look at the estimated probabilities of the model for flowers with petal widths varying from 0 to 3 cm.

X_new = np.linspace(-3, 3, 1000).reshape(-1, 1)
y_proba = log_reg.predict_proba(X_new)
decision_boundary = X_new[y_proba[:, 1] >= 0.5][0]

plt.figure(figsize = (8, 3), dpi = 200)
plt.plot(X_new, y_proba[:, 1], color = 'green', linestyle = '-', linewidth = 2, label = 'Iris virginica')
plt.plot(X_new, y_proba[:, 0], color = 'blue', linestyle = '--', linewidth = 1, label = 'Not Iris virginica')
plt.plot([decision_boundary, decision_boundary], [-1, 2], "k:", linewidth=2)

plt.scatter(X[y==0], y[y==0], marker = "s", color = 'blue')
plt.scatter(X[y==1], y[y==1], marker = '^', color = 'green')

plt.text(decision_boundary+0.02, 0.15, "Decision  boundary", fontsize=14, color="black", ha="center")
plt.arrow(decision_boundary, 0.08, -0.3, 0, head_width=0.05, head_length=0.1, fc='b', ec='b')
plt.arrow(decision_boundary, 0.92, 0.3, 0, head_width=0.05, head_length=0.1, fc='g', ec='g')

plt.xlabel("Petal width (cm)", fontsize=14)
plt.ylabel("Probability", fontsize=14)
plt.axis([-3, 3, -0.02, 1.02])
plt.legend(loc = 'best')
plt.show()
```

# Softmax Regression

```python
X = iris['data'][:, 2:4]  # petal length, petal width
y = iris['target']
X.shape
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(C = 10, solver='lbfgs', random_state = 42)
log_reg.fit(X, y)
```

```python
x0, x1 = np.meshgrid(np.linspace(0, 8, 500), np.linspace(0, 3.5, 200))
X_new = np.c_[x0.ravel(), x1.ravel()]

x0.shape, x1.shape
```

```python
y_proba = log_reg.predict_proba(X_new)
y_predict = log_reg.predict(X_new)

y_proba.shape
```

```python
zz1 = y_proba[:, 1].reshape(x0.shape)
zz = y_predict.reshape(x0.shape)

plt.figure(figsize=(12, 6))


from matplotlib.colors import ListedColormap
custom_cmap = ListedColormap(['#fafab0','#9898ff','#a0faa0'])
contour = plt.contour(x0, x1, zz1, cmap=plt.cm.brg)
plt.contourf(x0, x1, zz, cmap=custom_cmap)
plt.clabel(contour, inline=1, fontsize=12)

plt.scatter(X[y == 2, 0], X[y == 2, 1], marker = '^', color = 'green', label="Iris virginica")
plt.scatter(X[y == 1, 0], X[y == 1, 1], marker = 's', color = 'blue', label="Iris versicolor")
plt.scatter(X[y == 0, 0], X[y == 0, 1], marker = 'o', color = 'orange', label="Iris setosa")

plt.xlabel("Petal length", fontsize=14)
plt.ylabel("Petal width", fontsize=14)
plt.legend(loc="center left", fontsize=14)
plt.axis([0, 7, 0, 3.5])
plt.show()
```
